```python
# %matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# # Setting output cells to display all calculations
# from IPython.core.interactiveshell import InteractiveShell
# InteractiveShell.ast_node_interactivity = "all"

# File to Load 
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
df_city = pd.read_csv(city_data_to_load)
df_ride = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
df = df_city.merge(df_ride, how = 'outer', on ='city', left_index=True)


df['date'] = df['date'].apply(pd.to_datetime) # Just reformatting, No impact on below calculation


# Display the data table for preview
df.info()
```

# Four plots as part of one figure object

```python
#using Pivot Table to group the data for our scatter plot , (alternatively we can use groupby function )
df_pivot = df.pivot_table(values = ['fare', 'ride_id', 'driver_count'],   
                          index = ['city', 'type'],   
                          aggfunc = {'driver_count': np.mean, 'fare': np.mean, 'ride_id':len})

# remove city type from index, and set index to type
df_pivot.reset_index(inplace = True)
df_pivot.set_index('type', inplace = True)

df_pivot.head()


# Obtain the x and y coordinates for each of the three city types
x1 = df_pivot.loc['Urban','ride_id']
y1 = df_pivot.loc['Urban', 'fare']
s1 = df_pivot.loc['Urban', 'driver_count']
x2 = df_pivot.loc['Suburban','ride_id']
y2 = df_pivot.loc['Suburban', 'fare']
s2 = df_pivot.loc['Suburban', 'driver_count']
x3 = df_pivot.loc['Rural','ride_id']
y3 = df_pivot.loc['Rural', 'fare']
s3 = df_pivot.loc['Rural', 'driver_count']



# defining figure with 4 plots
fig, (ax1, ax2, ax3, ax4) = plt.subplots(nrows = 4, ncols =1, figsize =(8,40))

# FIRST PLOT : RIDE SHARING DATA (scatter plot of number of rides & average fare by type of cities)
#(parameter 's' is marker size for the plot->multiplied driver count by 10 to give a proper bubble size )
ax1.scatter(x1,y1, color='yellow', alpha=0.5, label = 'Urban', s = s1*10, linewidths=2, edgecolor='black')
ax1.scatter(x2,y2, color='green', alpha=0.5, label = 'Suburban', s = s2*10, linewidths=2, edgecolor='black')
ax1.scatter(x3,y3, color='blue', alpha=0.5, label = 'rural',s = s3*10, linewidths=2, edgecolor='black')
# Incorporate the other graph properties
ax1.set_xlabel('Total Number of Rides(per City)')
ax1.set_ylabel('Average Fare($)')
ax1.set_title('Pyber Ride Sharing Data', fontsize = 16)
ax1.grid()
# Create a legend
ax1.legend( title = 'City Types', loc='best')
# Incorporate a text label regarding circle size
ax1.text(x=10,y=16,s='Note:Circle size correlates with driver count per city, :', color='b')


## SECOND PLOT: TOTAL FARE BY CITY TYPE

# Calculate Type Percents
# creating df_pivot 2 for plotting next 2 pie charts
df_pivot2 = df.pivot_table(index = 'type', values =['fare','ride_id'], aggfunc={'fare':np.sum, 'ride_id':len})
df_pivot2.head()
df_pivot2.info()
fares = df_pivot2.fare # Series to plot
# Build Pie Chart
ax2.pie(x=fares, explode = (0, 0, 0.1),autopct='%1.1f%%' , labels =[ 'Rural', 'Suburban', 'Urban'])
ax2.set_title("Total Fare by City Type")


## 3rd PLOT: TOTAL RIDES BY CITY TYPE

# Calculate Ride Percents
rides = df_pivot2.ride_id # Series to plot
# Build Pie Chart
ax3.pie(x=rides, explode = (0, 0, 0.1),autopct='%1.1f%%' , labels =[ 'Rural', 'Suburban', 'Urban'])
ax3.set_title("Total Rides by City Type")

## 4th PLOT: TOTAL DRIVERS BY CITY TYPE

# Calculate driver Percents

df_pivot4 = df.pivot_table(index = ['type', 'city'], values ='driver_count', aggfunc={'driver_count': np.mean})
df_pivot4.head()
urban_drivers = df_pivot4.loc['Urban','driver_count'].sum()
suburban_drivers = df_pivot4.loc['Suburban','driver_count'].sum()
rural_drivers = df_pivot4.loc['Rural','driver_count'].sum()

drivers = [urban_drivers, suburban_drivers, rural_drivers]

# Build Pie Chart
ax4.pie(x=drivers, explode = (0.1, 0, 0),autopct='%1.1f%%' , labels =[ 'Urban', 'Suburban', 'Rural' ])
# Save Figure
ax4.set_title("Total Drivers by City Type")
fig.savefig("Results.pdf")
```
